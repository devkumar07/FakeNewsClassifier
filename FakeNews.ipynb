```python
from pandas import DataFrame
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('news.csv')

df = df.dropna(subset = ['title','label'], how = 'any')
df.shape
```

```python
x_train,x_test,y_train, y_test = train_test_split(df.title.values, df.label.values, test_size = 0.1, random_state = 7)

print(x_train)
print(y_train)
```

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform(x_train)

classifier = MultinomialNB()
targets = y_train 
classifier.fit(counts, targets)

x_test
```

```python
examples = ['Trump refuses to say whether he\'ll accept election results']
example_counts = vectorizer.transform(examples)
pred = classifier.predict(example_counts)
pred
```
